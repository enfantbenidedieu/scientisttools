```python
import pyreadr
import pandas as pd
result = pyreadr.read_r('./data/poison.rda')
poison = result["poison"]
poison.head(6)
```

```python
group_name = ["desc","desc2","symptom","eat"]
group = [2,2,5,6]
group_type = ["s","n","n","n"]
num_group_sup = [0,1]
```

```python
# Set variables weights in each groups
var_weights_mfa = {
    "symptom" : pd.Series([1/5]*5,index=["Nausea","Vomiting","Abdominals","Fever","Diarrhae"]),
    "eat" : pd.Series([1/6]*6,index=["Potato","Fish","Mayo","Courgette","Cheese","Icecream"])
}
var_weights_mfa
```

```python
from scientisttools import MFAQUAL

res_mfaqual = MFAQUAL(group=group,name_group=group_name,group_type=group_type,var_weights_mfa=None,num_group_sup=[0,1],parallelize=True)
res_mfaqual.fit(poison)
```
