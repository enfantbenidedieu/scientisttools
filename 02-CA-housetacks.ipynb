# CA - `housetasks` dataset

```python
#disable warnings
from warnings import simplefilter, filterwarnings
simplefilter(action='ignore', category=FutureWarning)
filterwarnings("ignore")
from great_tables import GT, html
def print_dt(data,title=None,subtitle=None,rowname=None,digits=4):
    dt = (GT(data=data.round(decimals=digits).rename_axis(rowname).reset_index())
          .tab_header(title=title, subtitle=subtitle))
    return dt
```

## `housetasks` dataset

```python
#housetasks dataset
from scientisttools.datasets import housetasks
(
    GT(housetasks.rename_axis("Rows").reset_index())
    .tab_header(title=html("<b>Housetasks Dataset</b>"))
)
```

## Instanciation and training

```python
#instanciation
from scientisttools import CA
res_ca = CA()
```

### `fit` function

```python
#fit function
res_ca.fit(housetasks)
```

### `fit_transform` function

```python
#fit_transform function
print_dt(res_ca.fit_transform(housetasks),rowname="Rows",title=html("<b>Rows coordinates</b>"))
```

### `inverse_transform` function

```python
#inverse_transform function
res_ca2 = CA(n_components=None).fit(housetasks)
print_dt(res_ca2.inverse_transform(res_ca2.row_.coord),rowname="Rows",title=html("<b>Children Dataset (Data reconstruction)</b>"))
```

### `transform` function

```python
#transform function
print_dt(res_ca.transform(res_ca.call_.X),rowname="Rows",title=html("<b>Rows coordinates</b>"))
```

## Row informations

```python
#rows informations
from scientisttools import get_ca_row
row = get_ca_row(res_ca)
row._fields
```

### Rows coordinates

```python
#rows coordinates
print_dt(row.coord,rowname="Rows",title=html("<b>Rows coordinates</b>"))
```

### Rows contributions

```python
#rows contributions
print_dt(row.contrib,rowname="Rows",title=html("<b>Rows contributions</b>"))
```

### Rows cos2

```python
#rows cos2
print_dt(row.cos2,rowname="Rows",title=html("<b>Rows cos2</b>"))
```

### Rows additionals informations

```python
#rows additionals informations
print_dt(row.cos2,rowname="Rows",title=html("<b>Rows additionals informations</b>"))
```

## Diagnostics tests - multivariate goodness of fit tests

```python
#multivariate goodness of fit
goodness = res_ca.goodness_
goodness._fields
```

### Pearson chi-squared test

```python
#pearson chi-squared test
print_dt(goodness.chi2,rowname="",title=html("<b>Pearson chi-squared test</b>"))
```

### Log-likelihood test (G-test)

```python
#log-likelihood test
print_dt(goodness.gtest,rowname="",title=html("<b>Log-likelihood (G-test) test</b>"))
```

### Others association test

```python
#others association tests
print_dt(goodness.association,rowname="",title=html("<b>Others association tests</b>"))
```
