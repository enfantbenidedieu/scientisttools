# FAMD - `wine` Dataset

```python
#disable warnings
from warnings import simplefilter, filterwarnings
simplefilter(action='ignore', category=FutureWarning)
filterwarnings("ignore")
from great_tables import GT, html
def print_dt(data,title=None,subtitle=None,rowname=None,digits=6):
    dt = (GT(data=data.round(digits).rename_axis(rowname).reset_index())
          .tab_header(title=title, subtitle=subtitle))
    return dt
```

## `wine` dataset

```python
#wine dataset
from scientisttools.datasets import wine
(
    GT(wine.rename_axis("Individuals").reset_index())
    .tab_header(title=html("<b>Wine Dataset</b>"))
)
```

## Instanciation & training

```python
#instanciation
from scientisttools import FAMD
res_famd = FAMD()
```

### `fit` function

```python
#fit function
res_famd.fit(wine)
```

### `fit_transform`

```python
#fit_transform
print_dt(res_famd.fit_transform(wine),rowname="Individuals",title=html("<b>Individuals coordinates</b>"))
```

### `transform`

```python
#transform
print_dt(res_famd.transform(res_famd.call_.X),rowname="Individuals",title=html("<b>Individuals coordinates</b>"))
```

## Eigen values

```python
#eigen values
from scientisttools import get_eig
print_dt(get_eig(res_famd),rowname="Dimensions",title=html("<b>Eigen values</b>"))
```

### Individuals additional informations

```python
#individuals additionals informations
print_dt(ind.infos,rowname="Individuals",title=html("<b>Individuals additionals informations</b>"))
```

### Individuals distance to origin
