# Multiple Factor Analysis (MFA)

```python
#changement de dossier
import os
os.chdir("d:/Bureau/PythonProject/packages/scientisttools/data/")

#importation des données
import pandas as pd
url = "http://factominer.free.fr/factomethods/datasets/wine.txt"
wine = pd.read_table(url,sep="\t")
wine.head()
```

```python
wine.info()
```

```python
df1 = wine.iloc[:,:2]
df1.head()
```

```python
df2 = wine.iloc[:,2:7]
df2.head()
```

```python
df3 = wine.iloc[:,7:10]
df3.head()
```

```python
df4 = wine.iloc[:,10:20]
df4.head()
```

```python
df5= wine.iloc[:,20:29]
df5.head()
```

```python
from scientisttools.decomposition import PCA

pca1 =PCA(normalize=True,n_components=3,row_labels=wine.index,col_labels=df2.columns).fit(df2)
eig1 = pca1.eig_[0]
eig1.round(2)
```

```python
pca2 =PCA(normalize=True,n_components=3,row_labels=wine.index,col_labels=df3.columns).fit(df3)
eig2 = pca2.eig_[0]
eig2.round(2)
```

```python
pca3 =PCA(normalize=True,n_components=3,row_labels=wine.index,col_labels=df4.columns).fit(df4)
eig3 = pca3.eig_[0]
eig3.round(2)
```

```python
pca4 =PCA(normalize=True,n_components=3,row_labels=wine.index,col_labels=df4.columns).fit(df5)
eig4 = pca4.eig_[0]
eig4.round(2)
```

### Pondération

#### Centrer et réduire les données

```python
from sklearn.preprocessing import StandardScaler

sc1 = StandardScaler(with_mean=True,with_std=True)
sc1.set_output(transform="pandas")
Z1 = sc1.fit_transform(df2)
Z1.head()
```

```python
sc2 = StandardScaler(with_mean=True,with_std=True)
sc2.set_output(transform="pandas")
Z2 = sc2.fit_transform(df3)
Z2.head()
```

```python
sc3 = StandardScaler(with_mean=True,with_std=True)
sc3.set_output(transform="pandas")
Z3 = sc3.fit_transform(df4)
Z3.head()
```

```python
sc4 = StandardScaler(with_mean=True,with_std=True)
sc4.set_output(transform="pandas")
Z4= sc2.fit_transform(df5)
Z4.head()
```

### Distance entre individus

```python
from scipy.spatial.distance import pdist,squareform
import numpy as np
Z =pd.concat([Z1,Z2,Z3,Z4],axis=1)
col_weight = np.append(np.append(np.repeat(eig1[0],Z1.shape[1]),np.repeat(eig2[0],Z2.shape[1])),
                       np.append(np.repeat(eig3[0],Z3.shape[1]),np.repeat(eig4[0],Z4.shape[1])))
col_weight
```

```python
row_dist = squareform(pdist(Z,metric= "seuclidean",V=col_weight)**2)
pd.DataFrame(row_dist)
```

#### Pondération des données centrées et réduites

```python
# Pondération
from mapply.mapply import mapply
Zb = pd.DataFrame().astype("float")
for eig, df in zip([eig1[0],eig2[0],eig3[0],eig4[0]],[Z1,Z2,Z3,Z4]):
    norm = mapply(df,lambda x : x/np.sqrt(eig),axis=0,progressbar=False,n_workers=-1)
    Zb = pd.concat([Zb,norm],axis=1)
```

## Mise en coeuvre de l'AFM

### Diagonalisation de la matrice

```python
pca = PCA(normalize=False,n_components=3,row_labels=Zb.index,col_labels=Zb.columns).fit(Zb)
eig = pca.eig_[0]
eig.round(2) 
```
