# HCPC with MCA

```python
import pandas as pd
import numpy as np
from scientisttools.clustering import HCPC
from scientisttools.pyplot import plot_hcpc_dendrogram,plotHCPC #,cluster_and_plot_dendrogram
from scientisttools.decomposition import MCA
```

```python
# Analyse factorielle des correspondances multiples
tea = pd.read_excel("./data/tea.xlsx")
display(tea.head())
```

```python
tea.shape
```

```python
mca = MCA(n_components=20,
          matrix_type="completed",
          var_labels=list(tea.columns[:18]),
          quanti_sup_labels=["age"],
          quali_sup_labels=tea.columns[19:],
          graph=False)

mca.fit(tea)
```

```python
from sklearn.pipeline import Pipeline
from sklearn.cluster import KMeans

skl = Pipeline([('scaler', MCA(n_components=20,
          matrix_type="completed",
          var_labels=list(tea.columns[:18]),
          quanti_sup_labels=["age"],
          quali_sup_labels=tea.columns[19:],
          graph=False)), 
          ('km', KMeans(n_clusters=3,random_state=0))])
clusters = skl.fit_predict(tea)
```

```python
pd.DataFrame(mca.mod_infos_[:,1])
```

```python
hcpc = HCPC(graph=False,n_clusters=3)
hcpc.fit(mca)
```
