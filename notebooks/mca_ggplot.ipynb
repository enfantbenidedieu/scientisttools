# MCA with plotnine

```python
# Import dataset
import pandas as pd
A= pd.read_excel("./scientisttools/data/races_canines_acm.xlsx",header=0,index_col=0)
```

```python
from scientisttools.decomposition import MCA
my_mca = MCA(n_components=None,
             row_labels=A.index[:27],
             var_labels=A.columns[:6],
             mod_labels=None,
             matrix_type="completed",
             benzecri=True,
             greenacre=True,
             row_sup_labels=A.index[27:],
             quali_sup_labels=["Fonction"],
             quanti_sup_labels=["Cote"],
             parallelize=False).fit(A)
```

## Individuals Factor Map - MCA

```python
from scientisttools.ggplot import fviz_mca_ind
p = fviz_mca_ind(my_mca,repel=True)
print(p)
```

```python
# Using cos2
p = fviz_mca_ind(my_mca,color="cos2",repel=True)
print(p)
```

```python
# using contrib
p = fviz_mca_ind(my_mca,color="contrib",repel=True)
print(p)
```

```python
import numpy as np
np.random.seed(123)
color_numpy = np.random.normal(size=27)
p = fviz_mca_ind(my_mca,color=color_numpy,repel=True)
print(p)
```

```python
# Using KMeans
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=np.random.seed(123), n_init="auto").fit(my_mca.row_coord_)
p = fviz_mca_ind(my_mca,color=kmeans,repel=True)
print(p)
```
