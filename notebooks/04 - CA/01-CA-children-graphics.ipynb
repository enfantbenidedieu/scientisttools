# Correspondence Analysis (CA) - `Children` Dataset graphs

```python
from warnings import simplefilter, filterwarnings
simplefilter(action='ignore', category=FutureWarning)
filterwarnings("ignore")
```

```python
from scientisttools import children, CA
#with supplementary columns
res_ca = CA(n_components=None,row_sup=(14,15,16,17),col_sup=(5,6,7),quali_sup=8)
res_ca.fit(children)
```

```python
#with supplementary quantitative variables
res_ca2 = CA(n_components=None,row_sup=(14,15,16,17),quanti_sup=(5,6,7),quali_sup=8)
res_ca2.fit(children)
```

```python
from scientisttools import fviz_eig
p = fviz_eig(res_ca,add_labels=True)
print(p)
```

## Row factor map

### Color by RGB color

```python
#color by RGB color
from scientisttools import fviz_ca_row
print(fviz_ca_row(res_ca,repel=True))
```

### Color by cos2

```python
#color by cos2
print(fviz_ca_row(res_ca,point_col_row="cos2",repel=True))
```

```python
from scientisttools import fviz_corrplot
print(fviz_corrplot(res_ca.row_.cos2,x_label="Rows points",title="Rows Cos2",legend_title="Cos2"))
```

```python
from scientisttools import fviz_cos2
print(fviz_cos2(res_ca,element="row",axis=0))
```

### Color by contrib

```python
#color by contrib
print(fviz_ca_row(res_ca,point_col_row="contrib",repel=True))
```

```python
print(fviz_corrplot(res_ca.row_.contrib,x_label="Rows points",title="Rows contributions",legend_title="Contrib"))
```

```python
from scientisttools import fviz_contrib
print(fviz_contrib(res_ca,element="row",axis=0))
```

### Color by quantitatives columns

```python
#color by quantitative variables
print(fviz_ca_row(res_ca,point_col_row="unqualified",repel=True))
```

### Color by numpy array

```python
#color using numpy array
import numpy as np
np.random.seed(123)
color_numpy = np.random.normal(size=res_ca.call_.X.shape[0])
print(fviz_ca_row(res_ca,point_col_row=color_numpy,repel=True))
```

### Color by Kmeans

```python
#color using kmeans clustering
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=np.random.seed(123), n_init="auto").fit(res_ca.row_.coord)
print(fviz_ca_row(res_ca,point_col_row=kmeans,repel=True,palette=("darkblue","darkred","darkgreen")))
```

### Color by qualitative

```python
#color using qualitative variable
print(fviz_ca_row(res_ca,repel=True,habillage="categorie",add_ellipses=False,palette=("darkblue","darkred","darkgreen","magenta")))
```

### Set limits

```python
p = fviz_ca_row(res_ca,repel=True,lim_cos2=0.9)
print(p)
```

```python
p = fviz_ca_row(res_ca,repel=True,lim_contrib=20)
print(p)
```
