# PCAMIX : predict & supvar

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
warnings.filterwarnings("ignore")
```

```python
from scientisttools import load_gironde
gironde = load_gironde()
```

```python
from scientisttools import splitmix
X_quant = splitmix(gironde)["quanti"]
X_qual = splitmix(gironde)["quali"]
```

## PCA with PCAMIX

```python
from scientisttools import PCAMIX
res_pca_pcamix = PCAMIX().fit(X_quant)
# Coordinates
res_pca_pcamix.ind_["coord"].head(6)
```

### transform

```python
# transform
res_pca_pcamix.transform(X_quant).head(6)
```

### predict

```python
from scientisttools import predictPCAMIX
predict = predictPCAMIX(res_pca_pcamix,X=X_quant)
predict.keys()
```

```python
# Coordinates
predict["coord"].head(6)
```

```python
# Square cosinus
predict["coord"].head(6)
```

```python
# Square distance
predict["dist"].head(6)
```

### supvar

```python
from scientisttools import supvarPCAMIX
supvar = supvarPCAMIX(res_pca_pcamix,X_quanti_sup=X_quant, X_quali_sup=X_qual)
supvar.keys()
```

#### Supplementary quantitatives

```python
supvar["quanti"].keys()
```

```python
# Coordinates
supvar["quanti"]["coord"].head(6)
```

```python
# Square cosinus
supvar["quanti"]["cos2"].head(6)
```

#### Supplementary qualitatives

```python
supvar["quali"].keys()
```

```python
# Coordinates
supvar["quali"]["coord"].head(6)
```

```python
# Square cosinus
supvar["quali"]["cos2"].head(6)
```

```python
# value-test
supvar["quali"]["vtest"].head(6)
```

```python
# Square distance
supvar["quali"]["dist"].head(6)
```

```python
# Square correlation ratio
supvar["quali"]["eta2"].head(6)
```
