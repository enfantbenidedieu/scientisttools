# Principal Components Analysis (PCA) - Decathlon Dataset graphs

```python
#disable warnings
from warnings import simplefilter, filterwarnings
simplefilter(action='ignore', category=FutureWarning)
filterwarnings("ignore")
#interactive notebook
from itables import init_notebook_mode
init_notebook_mode(all_interactive=True)
```

```python
from scientisttools import decathlon
decathlon
```

```python
from scientisttools import PCA
#instanciation
res_pca = PCA(ind_sup=(41,42,43,44,45),quanti_sup=(10,11),quali_sup=12,rotate=None)
res_pca.fit(decathlon)
```

## Eigenvalues

### Absolute

```python
from scientisttools import fviz_screeplot
p = fviz_screeplot(res_pca,choice="eigenvalue",add_labels=True)
print(p)
```

### Propoirtion

```python
# Proportion
p = fviz_screeplot(res_pca,choice="proportion",add_labels=True)
print(p)
```

### Cumulative

```python
# Cumulative
p = fviz_screeplot(res_pca,choice="cumulative",add_labels=True)
print(p)
```

## Individuals Factor Map

### Colors bu RGB colors

```python
from scientisttools import fviz_pca_ind
p = fviz_pca_ind(res_pca,repel=True)
print(p)
```

### Color by cos2

```python
p = fviz_pca_ind(res_pca,col_ind="cos2",repel=True)
print(p)
```

### Color by contrib

```python
p = fviz_pca_ind(res_pca,col_ind="contrib",repel=True)
print(p)
```

### Color by quantitative columns

```python
p = fviz_pca_ind(res_pca,col_ind=decathlon.columns[0],repel=True)
print(p)
```

### Color by qualitatives columns

```python
p = fviz_pca_ind(res_pca,habillage=decathlon.columns[-1],repel=True)
print(p)
```

```python
p = fviz_pca_ind(res_pca,shape_ind=("o",">"),habillage=decathlon.columns[-1],repel=True)
print(p)
```

### Color by numpy array

```python
import numpy as np
np.random.seed(123)
color_numpy = np.random.normal(size=res_pca.call_.X.shape[0])
p = fviz_pca_ind(res_pca,col_ind=color_numpy,repel=True)
print(p)
```

### Color by kmeans clustering

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, random_state=np.random.seed(123), n_init="auto").fit(res_pca.ind_.coord)
p = fviz_pca_ind(res_pca,col_ind=kmeans,repel=True)
print(p)
```

### Set limits

#### Limit with cos2

```python
p = fviz_pca_ind(res_pca,lim_cos2=0.8,repel=True)
print(p)
```

#### Limit with contrib

```python
p = fviz_pca_ind(res_pca,lim_contrib=10,repel=True)
print(p)
```
