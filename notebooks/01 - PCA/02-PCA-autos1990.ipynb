# Instanciation

```python
from scientisttools import PCA
res_pca = PCA(quali_sup=(6,7,8))
```

### `fit`

```python
#fit
res_pca.fit(autos1990)
```

### `fit_transform`

```python
#fit_transform
res_pca.fit_transform(autos1990)
```

### `inverse_transform`

```python
#inverse_transform
res_pca.inverse_transform(res_pca.ind_.coord)
```

### `transform`

```python
#transform
res_pca.transform(res_pca.call_.X)
```

#### Variables Cos2

```python
# Using kmeans
from sklearn.cluster import KMeans
import numpy as np
kmeans = KMeans(n_clusters=2, random_state=np.random.seed(123), n_init="auto").fit(var.coord)
kmeans.labels_
```
