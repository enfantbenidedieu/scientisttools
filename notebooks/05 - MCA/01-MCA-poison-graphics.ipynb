# MCA poison - graphics

```python
#disable warnings
from warnings import simplefilter, filterwarnings
simplefilter(action='ignore', category=FutureWarning)
filterwarnings("ignore")
```

```python
from scientisttools import poison, MCA
res_mca = MCA(quali_sup = (2,3),quanti_sup =(0,1))
res_mca.fit(poison)
```

## Eigenvalues

### Eigenvalue

```python
from scientisttools import fviz_screeplot
p = fviz_screeplot(res_mca,element="eigenvalue")
print(p)
```

### Proportion

```python
p = fviz_screeplot(res_mca,element="proportion")
print(p)
```

### Cumulative

```python
p = fviz_screeplot(res_mca,element="cumulative")
print(p)
```

## Individuals

### Color by RGB

```python
from scientisttools import fviz_mca_ind
p = fviz_mca_ind(res_mca,repel=True)
print(p)
```

### Color by cos2

```python
p = fviz_mca_ind(res_mca,col_ind="cos2",repel=True)
print(p)
```

```python
from scientisttools import fviz_corrplot
p = fviz_corrplot(res_mca.ind_.cos2,x_label="Individuals",legend_title="Cos2",title="Individuals Cos2")
print(p)
```

```python
from scientisttools import fviz_cos2
print(fviz_cos2(res_mca,element="ind"))
```

### Color by contrib

```python
p = fviz_mca_ind(res_mca,col_ind="contrib",repel=True)
print(p)
```

```python
p = fviz_corrplot(res_mca.ind_.contrib,x_label="Individuals",legend_title="Contrib",title="Individuals Contributions")
print(p)
```

```python
from scientisttools import fviz_contrib
print(fviz_contrib(res_mca,element="ind"))
```

## Color by quantitatives variables

```python
p = fviz_mca_ind(res_mca,col_ind="Age",repel=True)
print(p)
```

### Color by active qualitatives columns

```python
p = fviz_mca_ind(res_mca,habillage=poison.columns.tolist()[4],repel=True)
print(p)
```

### Color by supplementary qualitatives columns

```python
p = fviz_mca_ind(res_mca,habillage=poison.columns.tolist()[2],repel=True)
print(p)
```

### Color by numpy array

```python
import numpy as np
np.random.seed(123)
color_numpy = np.random.normal(size=res_mca.call_.X.shape[0])
p = fviz_mca_ind(res_mca,col_ind=color_numpy,repel=True)
print(p)
```

### Color by kmeans clustering

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=np.random.seed(123), n_init="auto").fit(res_mca.ind_.coord)
p = fviz_mca_ind(res_mca,col_ind=kmeans,repel=True)
print(p)
```

### set limits

#### limit cos2

```python
p = fviz_mca_ind(res_mca,lim_cos2=0.5,repel=True)
print(p)
```

#### limit contrib

```python
p = fviz_mca_ind(res_mca,lim_contrib=20,repel=True)
print(p)
```
