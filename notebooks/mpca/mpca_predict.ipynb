# MPCA : predict & supvar 

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
warnings.filterwarnings("ignore")
```

```python
from scientisttools import load_carsacpm
cars = load_carsacpm()
```

```python
from scientisttools import MPCA

res_mpca = MPCA(n_components=5)
res_mpca.fit(cars)
```

```python
# Individuals coordinates
res_mpca.ind_["coord"].head(6)
```

## transform

```python
# transform
res_mpca.transform(cars).head(6)
```

## predict

```python
from scientisttools import predictMPCA
predict = predictMPCA(res_mpca,X=cars)
predict.keys()
```

```python
# Coordinates
predict["coord"].head(6)
```

```python
# Square cosinus
predict["cos2"].head(6)
```

```python
# Square distance
predict["dist"].head(6)
```

## supvar

```python
# supvar
from scientisttools import splitmix, supvarMPCA
X_quant = splitmix(cars)["quanti"]
X_qual = splitmix(cars)["quali"]
supvar = supvarMPCA(res_mpca, X_quanti_sup=X_quant,X_quali_sup=X_qual)
supvar.keys()
```

### Supplementary quantitatives

```python
# Coordinates
supvar["quanti"]["coord"]
```

```python
# Square cosinus
supvar["quanti"]["cos2"]
```
