# CA with Matplotlib

```python
import pandas as pd
url = "http://factominer.free.fr/factomethods/datasets/women_work.txt"
women_work = pd.read_table(url,header=0)
```

```python
# Modèle avec colonnes supplémentaires
from scientisttools.decomposition import CA
my_ca = CA(n_components=None,
            row_labels=women_work.index.values,
            col_labels=women_work.columns[:3].values,
            row_sup_labels=None,
            col_sup_labels=women_work.columns[3:].values).fit(women_work)
```

## Row points - CA

```python
import matplotlib.pyplot as plt
from scientisttools.pyplot import plot_ca_row
fig, axe = plt.subplots(figsize=(8,8))
plot_ca_row(my_ca,axis=(0,1),repel=True)
plt.show()
```

```python
fig, axe = plt.subplots(figsize=(8,8))
plot_ca_row(my_ca,axis=(0,1),color="cos2",repel=True)
plt.show()
```

```python
fig, axe = plt.subplots(figsize=(8,8))
plot_ca_row(my_ca,axis=(0,1),color="contrib",repel=True)
plt.show()
```

```python
# Using  numpy color
import numpy as np
np.random.seed(123)
color_numpy = np.random.normal(size=3)
color_numpy
```

```python
fig, axe = plt.subplots(figsize=(8,8))
plot_ca_row(my_ca,axis=(0,1),color=color_numpy,repel=True)
plt.show()
```

```python
# Using KMeans
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=np.random.seed(123), n_init="auto").fit(my_ca.row_coord_)
kmeans.labels_
```

```python
fig, axe = plt.subplots(figsize=(8,8))
plot_ca_row(my_ca,axis=(0,1),color=kmeans,repel=True)
plt.show()
```
