```python
import os
import pandas as pd
from scientisttools.decomposition import FAMD

os.chdir("d:/Bureau/PythonProject/packages/scientisttools/scientisttools/data/")
autos = pd.read_excel("autos2005.xls",index_col=0)
```

```python
#instanciation
afdm = FAMD(n_components = None,
            row_labels=list(autos.index),
            row_sup_labels=None,
            quanti_labels=list(autos.columns[:10]),
            quanti_sup_labels=["CO2","prix"],
            quali_labels=["origine","Carburant"],
            quali_sup_labels=["type4X4"],
            parallelize=False).fit(autos)
```
