# Partial Principal Components Analysis (PartialPCA)  - Cars Dataset

```python
#disable warnings
from warnings import simplefilter, filterwarnings
simplefilter(action='ignore', category=FutureWarning)
filterwarnings("ignore")
from itables import init_notebook_mode
init_notebook_mode(all_interactive=True)
```

```python
# Load cars2006 datasets
from scientisttools import autos2006
autos2006
```

```python
# Chargement des librairies
from scientisttools import PartialPCA, fviz_screeplot,fviz_partialpca_ind
res_ppca = PartialPCA(partial=0,ind_sup=(18,19),quanti_sup=(6,7),quali_sup=8)
res_ppca.fit(autos2006)
```

```python
from scientisttools import summaryPartialPCA
summaryPartialPCA(res_ppca)
```

## Eigenvalues

### `absolute`

```python
p = fviz_screeplot(res_ppca,choice="eigenvalue",add_labels=True)
print(p)
```

### `percentage`

```python
p = fviz_screeplot(res_ppca,choice="proportion",add_labels=True)
print(p)
```

### `cumulative`

```python
p = fviz_screeplot(res_ppca,choice="cumulative",add_labels=True)
print(p)
```

### Color by quantitative columns

```python
p = fviz_partialpca_ind(res_ppca,col_ind="PUISS",repel=True,ind_sup=False,quali_sup=False)
print(p)
```

### Color by numpy array

```python
import numpy as np
np.random.seed(123)
color_numpy = np.random.normal(size=res_ppca.call_.X.shape[0])
p = fviz_partialpca_ind(res_ppca,col_ind=color_numpy,repel=True,ind_sup=False,quali_sup=False)
print(p)
```

### Color by kmeans clustering

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=np.random.seed(123), n_init="auto").fit(res_ppca.ind_.coord)
p = fviz_partialpca_ind(res_ppca,col_ind=kmeans,repel=True,ind_sup=False,quali_sup=False,palette=("blue","red","darkgreen"))
print(p)
```

### Color by qualitative variables

```python
p = fviz_partialpca_ind(res_ppca,habillage="FINITION",repel=True,ind_sup=False,quali_sup=False,palette=("blue","red","darkgreen"))
print(p)
```

### Set limit

```python
p = fviz_partialpca_ind(res_ppca,lim_cos2=0.8,repel=True,ind_sup=False,quali_sup=False)
print(p)
```

```python
p = fviz_partialpca_ind(res_ppca,lim_contrib=10,repel=True,ind_sup=False,quali_sup=False)
print(p)
```
