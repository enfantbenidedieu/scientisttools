# Factor Analysis of Mixed Data (FAMD) - `autos2005` Dataset

```python
from warnings import simplefilter, filterwarnings
simplefilter(action='ignore', category=FutureWarning)
filterwarnings("ignore")
from great_tables import GT, html
def print_dt(data,title=None,subtitle=None,rowname=None,digits=6):
    dt = (GT(data=data.round(digits).rename_axis(rowname).reset_index())
          .tab_header(title=title, subtitle=subtitle))
    return dt
```

## `autos2005` dataset

```python
#autos2005 dataset
from scientisttools import autos2005
(
    GT(autos2005.rename_axis("Individuals").reset_index())
    .tab_header(title=html("<b>Autos2005 Dataset</b>"))
    .tab_spanner(label=html("<b>Individuals</b>"),columns="Individuals")
    .tab_spanner(label=html("<b>Active<br>variables</b>"),columns=autos2005.columns.tolist()[:12])
    .tab_spanner(label=html("<b>Supp. <br>quantitatives</b>"),columns=autos2005.columns.tolist()[12:15])
    .tab_spanner(label=html("<b>Supp. <br>qualitative</b>"),columns=autos2005.columns.tolist()[15:])
)
```

## Instanciation and training

```python
#instanciation
from scientisttools import FAMD
res_famd = FAMD(ind_sup=(38,39,40,41,42,43,44),quanti_sup=(12,13,14),quali_sup=15)
res_famd.fit(autos2005)
```

## Others functions

### `fit_transform`

```python
#fit_transform
print_dt(res_famd.fit_transform(autos2005),rowname="Individuals",title=html("<b>Individuals coordinates</b>"))
```

### `transform`

```python
#transform
print_dt(res_famd.transform(res_famd.call_.X),rowname="Individuals",title=html("<b>Individuals coordinates</b>"))
```
