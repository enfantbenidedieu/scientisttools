# PCA -  `decathlon` dataset (`predictPCA` & `supvarPCA`)

```python
#disable warnings
from warnings import simplefilter, filterwarnings
simplefilter(action='ignore', category=FutureWarning)
filterwarnings("ignore")
from great_tables import GT, html
def print_dt(data,title=None,subtitle=None,rowname=None,digits=4):
    dt = (GT(data=data.round(digits).rename_axis(rowname).reset_index())
          .tab_header(title=title, subtitle=subtitle))
    return dt
```

## `decathlon` dataset

```python
#decathlon dataset
from scientisttools.datasets import load_decathlon
decathlon = load_decathlon(element="actif")
(
    GT(decathlon.rename_axis("Individuals").reset_index())
    .tab_header(title=html("<b>Decathlon Dataset (active elements)</b>"))
)
```

## Instanciation & training

```python
#instanciation
from scientisttools import PCA, predictPCA, supvarPCA
res_pca = PCA()
res_pca.fit(decathlon)
```

## Supplementary individuals

```python
#supplementary individuals
ind_sup = load_decathlon(element="ind_sup")
(
    GT(ind_sup.rename_axis("Individuals").reset_index())
    .tab_header(title=html("<b>Decathlon Dataset (supplementary individuals)</b>"))
)
```

## `transform`

```python
#supplementary individuals coordinates
print_dt(res_pca.transform(ind_sup),rowname="Individuals",title=html("<b>Supplementary individuals coordinates</b>"))
```

### `predictPCA`

```python
#predict on new rows
predict = predictPCA(res_pca,X=ind_sup)
predict._fields
```

### Supplementary individuals coordinates

```python
#supplementary individuals coordinates
print_dt(predict.coord,rowname="Individuals",title=html("<b>Supplementary individuals coordinates</b>"))
```

### Supplementary individuals cos2

```python
#supplementary individuals squared cosinus (cos2)
print_dt(predict.cos2,rowname="Individuals",title=html("<b>Supplementary individuals cos2</b>"))
```

### Supplementary individuals dist2

```python
#supplementary individuals dist2
print_dt(predict.dist2,rowname="Individuals",title=html("<b>Supplementary individuals dist2</b>"))
```

## `supvarPCA`

```python
#Supplementary variables
sup_var = load_decathlon(element="sup_var")
(
    GT(sup_var.rename_axis("Individuals").reset_index())
    .tab_header(title=html("<b>Decathlon Dataset (supplementary variables)</b>"))
    .tab_spanner(label=html("<b>Individuals</b>"),columns="Individuals")
    .tab_spanner(label=html("<b>Supplementary <br> quanti. var. </b>"),columns=sup_var.columns.tolist()[:2])
    .tab_spanner(label=html("<b>Supplementary <br> quali. var. </b>"),columns=sup_var.columns.tolist()[2:])
)
```

```python
#prediction on supplementary variables (quantitative and/or qualitatives)
sup_var_predict = supvarPCA(res_pca,X=sup_var)
```

### Supplementary quantitatives variables

```python
#supplementary quantitative variables informations
quanti_sup = sup_var_predict.quanti_sup
quanti_sup._fields
```

#### Supplementary quantitative variables coordinates

```python
#supplementary quantitative variables coordinates
print_dt(quanti_sup.coord,rowname="Variables",title=html("<b>Supplementary quantitative variables <br> coordinates</b>"))
```

#### Supplementary quantitative cos2

```python
#supplementary quantitative variables cos2
print_dt(quanti_sup.cos2,rowname="Individuals",title=html("<b>Supplementary quantitative variables cos2</b>"))
```

### Supplementary qualitative variables

```python
#supplementary qualitative variables informations
quali_sup = sup_var_predict.quali_sup
quali_sup._fields
```

#### Supplementary levels coordinates

```python
#supplementary levels coordinates
print_dt(quali_sup.coord,rowname="Levels",title=html("<b>Supplementary levels coordinates</b>"))
```

#### Supplementary levels vtest

```python
#supplementary levels vtest
print_dt(quali_sup.vtest,rowname="Levels",title=html("<b>Supplementary levels vtest</b>"))
```

#### Supplementary qualitative variables eta2

```python
#supplementary qualitative variables eta2
print_dt(quali_sup.eta2,rowname="Variables",title=html("<b>Supplementary qualitative variables eta2</b>"))
```

#### Supplementary levels cos2

```python
#supplementary levels cos2
print_dt(quali_sup.cos2,rowname="Levels",title=html("<b>Supplementary levels cos2</b>"))
```

#### Supplementary levels dist2

```python
#supplementary levels dist2
print_dt(quali_sup.dist2,rowname="Levels",title=html("<b>Supplementary <br> levels dist2</b>"))
```
