# MCA : predictMCA & supvarMCA

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
warnings.filterwarnings("ignore")
```

```python
from scientisttools import load_poison
poison = load_poison()
poison.head(6)
```

```python
from scientisttools import MCA
res_mca = MCA(n_components=5,quali_sup = [2,3],quanti_sup =[0,1])
res_mca.fit(poison)
```

```python
res_mca.quali_sup_["coord"]
```

```python
res_mca.quanti_sup_["coord"]
```

```python
from scientisttools import supvarMCA
X_quanti_sup = poison.iloc[:,[0,1]]
X_quali_sup =  poison.iloc[:,[2,3]]
supvar = supvarMCA(res_mca,X_quanti_sup=X_quanti_sup,X_quali_sup=X_quali_sup)
supvar["quanti"]["coord"]
```

```python
supvar["quali"]["coord"]
```

```python
res_mca.ind_["coord"].head(6)
```

```python
from scientisttools import predictMCA
predict = predictMCA(res_mca,poison.iloc[:,4:])
predict["coord"].head(6)
```
