### transform

```python
X = Biometrie1.copy()
for i in range(X.shape[1]):
    if X.iloc[:,i].dtype in ["object","category"]:
        value = {x : X.columns.tolist()[i]+"_"+str(x) for x in X.iloc[:,i].unique()}
        X.iloc[:,i] = X.iloc[:,i].map(value)
X
```

```python
res_mfamix.transform(X)
```
