```python
import pyreadr
import numpy as np
import pandas as pd
result = pyreadr.read_r('./data/mortality.rda') # also works for Rds, rda

# done! let's see what we got
# result is a dictionary where keys are the name of objects and the values python
# objects
print(result.keys()) 
```

```python
# Chargement des données
mortality = result["mortality"]
mortality.head(6)
```

```python
mortality.info()
```

```python
# Rename columns
group_cols1 = [("1979",col) for col in mortality.columns[:9].tolist()]
group_cols1
```

```python
group_cols2 = [("2006",col) for col in mortality.columns[9:].tolist()]
group_cols2
```

```python
columns = [*group_cols1,*group_cols2]
columns
```

```python
mortality.columns = pd.MultiIndex.from_tuples(columns)
mortality
```

```python
from scientisttools.decomposition import MFACT

res_mfact = MFACT(n_components=None,
                  row_labels=mortality.index.tolist(),
                  row_weight=None,
                  table_type="number",
                  group=["1979","2006"],
                  parallelize=True)
res_mfact.fit(mortality)
```
