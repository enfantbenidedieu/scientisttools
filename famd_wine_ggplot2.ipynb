# FAMD wine ggplot2

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
warnings.filterwarnings("ignore")
```

```python
#importation des données
import pandas as pd
url = "http://factominer.free.fr/factomethods/datasets/wine.txt"
wine = pd.read_table(url,sep="\t")
```

```python
from scientisttools import FAMD
res_famd = FAMD(n_components=5)
res_famd.fit(wine)
```

## Eigenvalues

```python
from scientisttools import fviz_screeplot
p = fviz_screeplot(res_famd)
print(p)
```

## Individuals Factor Map

### Color by RGB colors

```python
from scientisttools import fviz_famd_ind
p = fviz_famd_ind(res_famd,repel=True)
print(p)
```

### Color by Cos2

```python
p = fviz_famd_ind(res_famd,color="cos2",repel=True)
print(p)
```

```python
from scientisttools import fviz_corrplot
p = fviz_corrplot(res_famd.ind_["cos2"],colors=["#00AFBB", "#E7B800", "#FC4E07"])
print(p)
```

### Color by contrib

```python
p = fviz_famd_ind(res_famd,color="contrib",repel=True)
print(p)
```

### Color by quantitatives columns

```python
p = fviz_famd_ind(res_famd,color="Nuance",repel=True)
print(p)
```

### Color by actives qualitatives columns

```python
p = fviz_famd_ind(res_famd,habillage="Soil",repel=True)
print(p)
```

### Color by supplementary qualitatives columns

```python
p = fviz_famd_ind(res_famd,habillage="Label",repel=True,add_ellipses=True)
print(p)
```

### Color by numpy array

```python
import numpy as np
np.random.seed(123)
color_numpy = np.random.normal(size=res_famd.call_["X"].shape[0])
p = fviz_famd_ind(res_famd,color=color_numpy,repel=True)
print(p)
```

### Color by kmeans clustering

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=3, random_state=np.random.seed(123), n_init="auto").fit(res_famd.ind_["coord"])
p = fviz_famd_ind(res_famd,color=kmeans,repel=True)
print(p)
```

### Set limits

#### Limit cos2

```python
p = fviz_famd_ind(res_famd,lim_cos2=0.50,repel=True)
print(p)
```

#### limit contrib

```python
p = fviz_famd_ind(res_famd,lim_contrib=20,repel=True)
print(p)
```
