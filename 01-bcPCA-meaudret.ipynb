# bcPCA - `meaudret` dataset

```python
#disable warnings
from warnings import simplefilter, filterwarnings
simplefilter(action='ignore', category=FutureWarning)
filterwarnings("ignore")
from great_tables import GT, html
def print_dt(data,title=None,subtitle=None,rowname=None,digits=4):
    dt = (GT(data=data.round(digits).rename_axis(rowname).reset_index())
          .tab_header(title=title, subtitle=subtitle))
    return dt
```

## `meaudret` dataset

```python
from scientisttools.datasets import load_meaudret
meaudret = load_meaudret()
(
    GT(meaudret.rename_axis("Sites").reset_index())
    .tab_header(title=html("<b>Meaudret Dataset</b>"))
    .tab_spanner(label=html("<b>Sites</b>"),columns="Sites")
    .tab_spanner(label=html("<b>Dependent<br>variables</b>"),columns=meaudret.columns.tolist()[:9])
    .tab_spanner(label=html("<b>class <br>variable</b>"),columns=meaudret.columns.tolist()[9:10])
    .tab_spanner(label=html("<b>Supplementary <br>variable</b>"),columns=meaudret.columns.tolist()[10:])
)
```

## Instanciation & training

```python
#instanciation
from scientisttools import bcPCA
res_bcpca = bcPCA(group=9,sup_var=range(10,24))
```

### `fit` function

```python
#fit function
res_bcpca.fit(meaudret)
```

### `fit_transform` function

```python
#fit_transform function
print_dt(res_bcpca.fit_transform(meaudret).head(10),rowname="Individuals",title=html("<b>Individuals coordinates</b>"))
```

### `transform` function

```python
#transform function
print_dt(res_bcpca.transform(res_bcpca.call_.X).head(10),rowname="Individuals",title=html("<b>Individuals coordinates</b>"))
```

## Eigen values

```python
#eigenvalues
from scientisttools import get_eig
print_dt(get_eig(res_bcpca),rowname="Dimensions",title=html("<b>Eigen values</b>"))
```
