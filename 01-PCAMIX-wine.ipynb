# PCAMIX - `wine` dataset

```python
from warnings import simplefilter, filterwarnings
simplefilter(action='ignore', category=FutureWarning)
filterwarnings("ignore")
from great_tables import GT, html
def print_dt(data,title=None,subtitle=None,rowname=None,digits=4):
    dt = (GT(data=data.round(digits).rename_axis(rowname).reset_index())
          .tab_header(title=title, subtitle=subtitle))
    return dt
```

 ## `wine` dataset

```python
#wine dataset
from scientisttools.datasets import wine
(
    GT(wine.rename_axis("Individuals").reset_index())
    .tab_header(title=html("<b>Wine Dataset</b>"))
)
```

## Instanciation & training

```python
#instanciation
from scientisttools import PCAMIX
res_pcamix = PCAMIX()
```

### `fit`function

```python
#fit function
res_pcamix.fit(wine)
```

### `fit_transform` function

```python
#fit_transform function
print_dt(res_pcamix.fit_transform(wine),rowname="Individuals",title=html("<b>Individuals coordinates</b>"))
```

### `transform`function

```python
#transform function
print_dt(res_pcamix.transform(res_pcamix.call_.X),rowname="Individuals",title=html("<b>Individuals coordinates</b>"))
```

## Eigen values

```python
#eigen values
from scientisttools import get_eig
print_dt(get_eig(res_pcamix),rowname="Dimensions",title=html("<b>Eigen values</b>"))
```

### Individuals additionals informations

```python
#individuals additionals informations
print_dt(ind.infos,rowname="Individuals",title=html("<b>Individuals additionals informations</b>"))
```

### Diagnostics tests - multivariate goodness of fit

```python
#diagnostics tests - multivariate goodness of fit
goodness = res_pcamix.goodness_
goodness._fields
```

#### Pearson chi-squared tests

```python
#pearson chi-squared tests
print_dt(goodness.chi2,rowname="",title=html("<b>Pearson chi-squared tests</b>"))
```

#### Log-likelihood tests (G-test)

```python
#log-likelihood test (G-test)
print_dt(goodness.gtest,rowname="",title=html("<b>Log-likelihood tests (G-test)</b>"))
```

#### Others association tests

```python
#others association tests
print_dt(goodness.association,rowname="",title=html("<b>Others association tests</b>"))
```
