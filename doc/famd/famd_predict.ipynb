# FAMD : predict & supvar

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
warnings.filterwarnings("ignore")
```

```python
from scientisttools import load_gironde
gironde = load_gironde()
```

```python
from scientisttools import splitmix
X_quant = splitmix(gironde)["quanti"]
X_qual = splitmix(gironde)["quali"]
```

## PCA with FAMD

```python
from scientisttools import FAMD
res_pca_famd = FAMD().fit(X_quant)
# Coordinates
res_pca_famd.ind_["coord"].head(6)
```

### transform

```python
# transform
res_pca_famd.transform(X_quant).head(6)
```

### predict

```python
from scientisttools import predictFAMD
predict = predictFAMD(res_pca_famd,X=X_quant)
predict.keys()
```

```python
# Coordinates
predict["coord"].head(6)
```

```python
# Square cosinus
predict["cos2"].head(6)
```

```python
# Square distance
predict["dist"].head(6)
```

### supvar

```python
from scientisttools import supvarFAMD
supvar = supvarFAMD(res_pca_famd,X_quanti_sup=X_quant, X_quali_sup=X_qual)
supvar.keys()
```

#### Supplementary quantitatives

```python
supvar["quanti"].keys()
```

```python
# Coordinates
supvar["quanti"]["coord"].head(6)
```

```python
# Square cosinus
supvar["quanti"]["cos2"].head(6)
```

#### Supplementary qualitatives

```python
supvar["quali"].keys()
```

```python
# Coordinates
supvar["quali"]["coord"].head(6)
```

```python
# Square cosinus
supvar["quali"]["cos2"].head(6)
```

```python
# value-test
supvar["quali"]["vtest"].head(6)
```

```python
# Square distance
supvar["quali"]["dist"].head(6)
```
