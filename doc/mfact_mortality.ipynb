```python
import pyreadr
result = pyreadr.read_r('./data/mortality.rda') 

mortality = result["mortality"]
mortality.head(6)
```

```python
mortality.info()
```

```python
from scientisttools import MFACT

res_mfact = MFACT(group=[9,9],name_group=["1979","2006"],parallelize=True)
res_mfact.fit(mortality)
```
