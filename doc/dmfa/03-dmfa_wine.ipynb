# Dual Multiple Factor Analysis (DMFA) - Wine Dataset

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
warnings.filterwarnings("ignore")
#activation des tables interactifs
from itables import init_notebook_mode
init_notebook_mode(all_interactive=True)
```

```python
# load wine dataset
from scientisttools import load_wine
wine = load_wine()
wine
```

```python
# Dual Multiple Factor Analysis (DMFA)
from scientisttools import DMFA
res_dmfa = DMFA(num_fact=0,quali_sup=1,quanti_sup=[29,30])
res_dmfa.fit(wine)
```

## Eigenvalues

```python
#eigenvalues
from scientisttools import get_eig
get_eig(res_dmfa)
```

```python
#scree plot
from scientisttools import fviz_screeplot
p = fviz_screeplot(res_dmfa)
print(p)
```

## Individuals informations

```python
#individuals informations
from scientisttools import get_dmfa_ind
ind = get_dmfa_ind(res_dmfa)
ind._fields
```

### Individuals factor coordinates

```python
#individuals
ind.coord
```

```python
#individuals factor map
from scientisttools import fviz_dmfa_ind
p = fviz_dmfa_ind(res_dmfa,repel=True,quali_sup=False)
print(p)
```

### Individuals square cosinus

```python
#individuals square cosinus
ind.cos2
```

```python
#order individuals by square cosinus
import patchworklib as pw
from scientisttools import fviz_cos2
p1, p2 = fviz_cos2(res_dmfa,choice="ind",axis=0,top_cos2=10), fviz_cos2(res_dmfa,choice="ind",axis=1,top_cos2=10)
g1, g2 = pw.load_ggplot(p1, figsize=(4,4)), pw.load_ggplot(p2, figsize=(4,4))
(g1|g2).savefig()
```

### Individuals contributions

```python
#individuals contributions
ind.contrib
```

```python
#order individuals by contributions
from scientisttools import fviz_contrib
p1, p2 = fviz_contrib(res_dmfa,choice="ind",axis=0,top_contrib=10), fviz_contrib(res_dmfa,choice="ind",axis=1,top_contrib=10)
g1, g2 = pw.load_ggplot(p1, figsize=(4,4)), pw.load_ggplot(p2, figsize=(4,4))
(g1|g2).savefig()
```
