# Partial Principal Components Analysis (PartialPCA) - Body Dataset

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
warnings.filterwarnings("ignore")
#activation des tables interactifs
from itables import init_notebook_mode
init_notebook_mode(all_interactive=True)
```

```python
# Load dataset
from scientisttools import load_body
data = load_body()
data
```

```python
# Drop gender
body = data.drop(columns=["gender"])
body.columns = [x.replace(".","_") for x in body.columns]
```

```python
# Concatenate
import pandas as pd
sup = data.drop(columns=["weight","height"])
sup.columns = [x.replace(".","__") for x in sup.columns]
D = pd.concat((body,sup),axis=1)
D.info()
```

```python
from scientisttools import PartialPCA
res_partialpca = PartialPCA(standardize=True,partial=["weight","height"],quanti_sup=list(range(14,26)),quali_sup=26,parallelize=False)
res_partialpca.fit(D)
```
