# Principal Components Analysis (PCA) - Decathlon2 Dataset

```python
#suppress wranings and message
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
warnings.filterwarnings("ignore")
#activation des tables interactifs
from itables import init_notebook_mode 
init_notebook_mode(all_interactive=True)
```

```python
# Load deacthlon2 dataset
from scientisttools import load_decathlon2
decathlon = load_decathlon2()
decathlon
```

```python
# Instanciation of PCA
from scientisttools import PCA
res_pca = PCA(standardize=True,ind_sup=list(range(23,27)),quanti_sup=[10,11],quali_sup=12,parallelize=True)
res_pca.fit(decathlon)
```

## Eigenvalues

```python
# Extract eigenvalues/variances
from scientisttools import get_eigenvalue
eig_val = get_eigenvalue(res_pca)
eig_val
```

```python
# sceree plot
from scientisttools import fviz_eig
p = fviz_eig(res_pca,add_labels=True)
print(p)
```

### Square cosinus

```python
# Square cosinus
var.cos2
```

```python
#plot square cosinus
import patchworklib as pw
from scientisttools import fviz_cos2
p1, p2 = fviz_cos2(res_pca,choice="var",axis=0), fviz_cos2(res_pca,choice="var",axis=1)
g1, g2 = pw.load_ggplot(p1, figsize=(4,4)), pw.load_ggplot(p2, figsize=(4,4))
(g1|g2).savefig()
```

### Color by a random sample

```python
#color variables using random sample
import numpy as np 
np.random.seed(123)
my_cont_var = np.random.normal(size=10)
p = fviz_pca_var(res_pca,col_var=my_cont_var,gradient_cols = ("blue", "yellow", "red"),quanti_sup=False)
print(p)
```

### Color by kmeans group

```python
from sklearn.cluster import KMeans
kmeans = KMeans(n_clusters=2, random_state=np.random.seed(123), n_init="auto").fit(var.coord)
p = fviz_pca_var(res_pca,col_var=kmeans,palette = ["blue", "red"],legend_title="Cluster",quanti_sup=False)
print(p)
```
