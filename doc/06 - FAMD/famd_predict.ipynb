# Factor Analysis of Mixed Data (FAMD) - Gironde Dataset (predict & supvar)

```python
import warnings
from itables import init_notebook_mode
warnings.simplefilter(action='ignore', category=FutureWarning)
warnings.filterwarnings("ignore")
init_notebook_mode(all_interactive=True)
```

```python
#load gironde dataset
from scientisttools import load_gironde
gironde = load_gironde()
gironde
```

```python
#split dataset
from scientisttools import splitmix
X_quant, X_qual = splitmix(gironde).quanti,  splitmix(gironde).quali
```

## PCA with `FAMD` function

```python
from scientisttools import FAMD
res_pca_famd = FAMD(quali_sup=X_qual.columns.tolist()).fit(gironde)
# individuals Coordinates
res_pca_famd.ind_.coord
```

### transform

```python
# transform
res_pca_famd.transform(X_quant)
```

### predict

```python
from scientisttools import predictFAMD
predict = predictFAMD(res_pca_famd,X=X_quant)
predict._fields
```

#### Individuals factor coordinates

```python
#with FAMD functions
res_pca_famd.ind_.coord
```

```python
#with predict functions
predict.coord
```

#### Individuals square cosinus (cos2)

```python
#with FAMD functions
res_pca_famd.ind_.cos2
```

```python
# Square cosinus
predict.cos2
```

#### Individuals square distance to origin

```python
#with FAMD functions
res_pca_famd.ind_.infos
```

```python
# Square distance
predict.dist
```
