```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
warnings.filterwarnings("ignore")
```

```python
from scientisttools import load_gironde, splitmix
gironde = load_gironde()
X_quant = splitmix(gironde)["quanti"]
X_qual = splitmix(gironde)["quali"]
```

```python
from scientisttools import PCA, supvarPCA
res_pca = PCA().fit(X_quant)
```

## supvar

```python
# sup variables
supvar = supvarPCA(res_pca,X_quanti_sup=X_quant, X_quali_sup=X_qual)
supvar.keys()
```
