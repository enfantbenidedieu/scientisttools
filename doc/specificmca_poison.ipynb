```python
from scientisttools import load_poison
poison = load_poison()
poison.head(6)
```

```python
from scientisttools import SpecificMCA
# Junk 
excl = {"Sick" : "Sick_n", "Sex" : "F"}
res_mca = SpecificMCA(n_components=5,excl=excl,quanti_sup=[0,1])
res_mca.fit(poison)
```
