# MFAMIX - predict & supvar

```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
warnings.filterwarnings("ignore")
```

```python
from scientisttools import load_gironde
gironde = load_gironde()
gironde.info()
```

```python
# Fill NA by mean in income
gironde.loc[:,"income"] = gironde.loc[:,"income"].fillna(gironde.loc[:,"income"].mean())
```

```python
from scientisttools import MFAMIX
name = ["employment","housing","services","environment"] 
group_type = ["s","m","n","s"]
res_mfamix = MFAMIX(n_components=5,group=[9,5,9,4],name_group=name,group_type=group_type,ind_sup=list(range(500,542)))
res_mfamix.fit(gironde)
```

## predict function

```python
import pandas as pd
ind_sup_index = list(range(500,542))
ind_sup = gironde.iloc[ind_sup_index,:]
from scientisttools import predictMFA
predict = predictMFA(res_mfamix,X=ind_sup)
predict.keys()
```

## supvar function

```python
from scientisttools import supvarMFA
X_group_sup = gironde.iloc[:500,:]
X_group_sup.columns = [x+"-2" for x in X_group_sup.columns]
name2 = ["employment2","housing2","services2","environment2"] 
supvar = supvarMFA(res_mfamix,X_group_sup=X_group_sup,group_sup=[9,5,9,4],name_group_sup=name2,group_sup_type=group_type)
supvar.keys()
```
