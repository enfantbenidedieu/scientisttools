## MPCA with supplementary individuals

```python
# Concatenate
cars3 = cars.copy()
cars3.index = [x+"2" for x in cars3.index]
Cars2 = pd.concat((cars,cars3),axis=0)
Cars2.info()
```

```python
res_mpca3 = MPCA(n_components=5,ind_sup=list(range(cars.shape[0],Cars2.shape[0])))
res_mpca3.fit(Cars2)
```

### Supplementary individuals
