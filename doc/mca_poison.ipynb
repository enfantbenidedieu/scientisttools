```python
import warnings
warnings.simplefilter(action='ignore', category=FutureWarning)
warnings.filterwarnings("ignore")
```

```python
import pyreadr
result = pyreadr.read_r('./data/poison.rda')
poison = result["poison"]
poison.head(6)
```

```python
from scientisttools import MCA
res_mca = MCA(n_components=5,ind_sup=list(range(50,55)),quali_sup = [2,3],quanti_sup =[0,1])
res_mca.fit(poison)
```

```python
res_mca.fit_transform(poison).iloc[:6,:5]
```
