# Discriminant Analysis with categorical variables (DISQUAL)

```python
# Chargement des librairies
import numpy as np
import pandas as pd

DTrain = pd.read_excel("./data/CongressVotePipeline.xlsx",sheet_name="train",header=0)
display(DTrain.head())
```

```python
DTrain.info()
```

```python
from scientisttools.discriminant_analysis import DISQUAL
```

```python
disqual = DISQUAL(n_components=None,
                  target=["group"],
                  features_labels=list(DTrain.columns[:-1]),
                  row_labels=DTrain.index)
disqual.fit(DTrain)
```

## Résulatats de l'ACM

### Valeurs propres

```python
eig = disqual.eig_
display(eig)
```

```python
mod = disqual.mod_
mod_infos = mod["infos"]
display(mod_infos)
```

```python
mod_coord = mod["coord"]
display(mod_coord)
```

```python
# Fonction de projection
fproj = disqual.projection_function_
display(fproj)
```

```python
# Coordonnées des individus
row = disqual.row_
row_coord = row["coord"]
display(row_coord.head(10))
```

```python
# Coeffcients du lDA
lda_coef = disqual.lda_coef_
display(lda_coef)
```

```python
lda_intercept = disqual.lda_intercept_
display(lda_intercept)
```

```python
# Evaluation globale
se = disqual.statistical_evaluation_
display(se)
```

```python
coef = disqual.coef_
display(coef)
```

```python
intercept = disqual.intercept_
display(intercept)
```

```python
DTest = pd.read_excel("./data/CongressVotePipeline.xlsx",sheet_name="test",header=0)
DTest.head()
```

```python
DTest.info()
```

```python
XTest = DTest[DTest.columns[:-1]]
new_coord =  disqual.transform(XTest)
new_coord.iloc[:,:2]
```

```python
XTrain = DTrain[DTrain.columns[:-1]]
coord = disqual.transform(XTrain)
coord.iloc[:,:2]
```

```python
pred_train = disqual.predict(XTrain)
pred_train
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix(DTrain[DTrain.columns[-1]],pred_train)
```

```python
pred_test = disqual.predict(XTest)
pred_test
```

```python
confusion_matrix(DTest[DTest.columns[-1]],pred_test)
```

```python
prob_train = disqual.predict_proba(XTrain)
prob_train
```

```python
prob_test = disqual.predict_proba(XTest)
prob_test
```
