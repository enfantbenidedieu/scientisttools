# Discriminant Analysis with categorical variables (DISCQUAL)

```python
# Chargement des librairies
import numpy as np
import pandas as pd

vote = pd.read_excel("./data/Data_Illustration_Livre_ADL.xlsx",sheet_name="VOTE",header=0)
vote.head()
```

```python
vote = vote[["group","adop_budget","edu_spending"]]
vote.head()
```

```python
X = vote[["adop_budget","edu_spending"]]
X.head()
```

```python
from scientisttools.discriminant_analysis import DISCQUAL
```

```python
disqual = DISCQUAL(n_components=None,
                    target=["group"],
                    features_labels=["adop_budget","edu_spending"],
                    row_labels=X.index)
disqual.fit(vote)
```

```python
pd.DataFrame(disqual.mod_coord_)
```
