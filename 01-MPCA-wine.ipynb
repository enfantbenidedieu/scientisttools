# MPCA - `wine` dataset

```python
#disable warnings
from warnings import simplefilter, filterwarnings
simplefilter(action='ignore', category=FutureWarning)
filterwarnings("ignore")
from great_tables import GT, html
def print_dt(data,title=None,subtitle=None,rowname=None,digits=4):
    dt = (GT(data=data.round(digits).rename_axis(rowname).reset_index())
          .tab_header(title=title, subtitle=subtitle))
    return dt
```

## `wine` dataset

```python
#wine dataset
from scientisttools.datasets import wine
(
    GT(wine.rename_axis("Individuals").reset_index())
    .tab_header(title=html("<b>Wine Dataset</b>"))
)
```

## Instanciation & training

```python
from scientisttools import MPCA
res_mpca = MPCA()
```

### `fit` function

```python
#fit function
res_mpca.fit(wine)
```

## Eigen values

```python
#eigen values
from scientisttools import get_eig
print_dt(get_eig(res_mpca),rowname="Dimensions",title=html("<b>Eigen values</b>"))
```
