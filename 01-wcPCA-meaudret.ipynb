# Within-class Principal Component Analysis (WithinPCA) - `meaudret` dataset

```python
#disable warnings
from warnings import simplefilter, filterwarnings
simplefilter(action='ignore', category=FutureWarning)
filterwarnings("ignore")
from great_tables import GT, html
def print_dt(data,title=None,subtitle=None,rowname=None,digits=4):
    dt = (GT(data=data.round(digits).rename_axis(rowname).reset_index())
          .tab_header(title=title, subtitle=subtitle))
    return dt
```

## `meauret` dataset

```python
from scientisttools.datasets import meaudret
(
    GT(meaudret.rename_axis("Sites").reset_index())
    .tab_header(title=html("<b>Meaudret Dataset</b>"))
    .tab_spanner(label=html("<b>Sites</b>"),columns="Sites")
    .tab_spanner(label=html("<b>Dependent<br>variables</b>"),columns=meaudret.columns.tolist()[:9])
    .tab_spanner(label=html("<b>group <br>variable</b>"),columns=meaudret.columns.tolist()[9])
    .tab_spanner(label=html("<b>Supplementary <br> variables</b>"),columns=meaudret.columns.tolist()[10:])
)
```

## Instanciation and training

```python
from scientisttools import wcPCA
res_wcpca = wcPCA(group=9,sup_var=range(10,24))
```

### `fit` function

```python
#fit function
res_wcpca.fit(meaudret)
```

### `fit_transform` function

```python
#fit_transform
print_dt(res_wcpca.fit_transform(meaudret).head(10),rowname="Individuals",title=html("<b>Individuals coordinates</b>"))
```

### `transform` function

```python
#transform
print_dt(res_wcpca.transform(res_wcpca.call_.X).head(10),rowname="Individuals",title=html("<b>Individuals coordinates</b>"))
```

### `inverse_transform` function

```python
#with max components
from pandas import concat
res_wcpca2 = wcPCA(n_components=None,group=9,sup_var=range(10,24))
res_wcpca2.fit(meaudret)
X = concat((res_wcpca2.ind_.coord,meaudret.iloc[:,9]),axis=1)
print_dt(res_wcpca2.inverse_transform(X).head(10),rowname="Individuals",title=html("<b>Active dataset</b>"))
```
