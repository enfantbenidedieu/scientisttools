```python
from scientisttools import load_mortality
mortality = load_mortality()
mortality.shape
```

```python
from scientisttools import MFACT
res_mfact = MFACT(group=[9,9],name_group=["1979","2006"],row_sup=list(range(50,mortality.shape[0])),parallelize=True)
res_mfact.fit(mortality)
```

## predict function

```python
from scientisttools import predictMFACT
row_sup = mortality.iloc[list(range(50,mortality.shape[0])),:]
```
